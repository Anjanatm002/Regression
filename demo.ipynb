Simple linear regression is used when the relationship between the independent variable and the dependent variable is linear. In this case, the model is a linear equation of the form: Y = a + bX, where Y is the dependent variable , X is the independent variable, a is the intercept or constant, and b is the slope of the line. 

Multiple linear regression is used when there are multiple independent variables. In this case, the model is a linear equation of the form
Y = a + b1X1 + b2X2 + ... + bkXk, where Y is the dependent variable, X1, X2, ..., Xk are the independent variables, and a, b1, b2, ..., bk are the coefficients of the model.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
Area=[2600,2800,3200,3500,4000]
Price=[5500000,5850000,6100000,6700000,7200000]
x_find=3800

# Calculate the regression line
z = np.polyfit(Area, Price,1)
p = np.poly1d(z)


plt.scatter(Area,Price)
plt.plot(Area, p(Area), "r--")

plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.grid(True)
plt.show()

# Print the regression equation
print("Regression Equation: y = {:.2f}x + {:.2f}".format(z[0], z[1]))
```
