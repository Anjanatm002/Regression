```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
sns.regplot(x=df["R&D Spend"],y=y,color="r")
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
s=make_column_transformer((OneHotEncoder(handle_unknown='ignore' ) , ['State' ] ) , remainder='passthrough')
x=s.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,xtest,y_train,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train
```

```python
xtest
```

```python
y_train
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(xtest)
y_pred
```

```python
#Data Frame
df1=pd. DataFrame({'Actual_value':ytest,'Predicted_value':y_pred,'Difference':ytest-y_pred})
df1
```

```python
#Performance Evaluation Methods
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error is",mean_absolute_error(ytest, y_pred) )
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("Mean absolute Percentage Error is",mean_absolute_percentage_error(ytest,y_pred) )
mse=mean_absolute_percentage_error(ytest,y_pred)
```

```python
print("RMSE is",np.sqrt(mse))
```
